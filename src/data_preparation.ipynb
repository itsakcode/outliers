```python
import pandas as pd
```

```python
# Define chunk size (number of rows to read at a time)
chunk_size = 10000

# Initialize an empty list to store chunks of Dataframes
list_of_df = []
```

```python
# Read the CSV file in chunks
for chunk in pd.read_csv("../Datasets/ZILLOW_DATA.csv", chunksize=chunk_size):
    # add each dataframe to list of dataframes
    list_of_df.append(chunk)

# Concatenate all chunks into a single DataFrame
zillow_data_df = pd.concat(list_of_df, ignore_index=True)
zillow_data_df
```

```python
# Initialize an empty list to store chunks of Dataframes
list_of_df = []

# Read the CSV file in chunks
for chunk in pd.read_csv("../Datasets/ZILLOW_INDICATORS.csv", chunksize=chunk_size):
    # add each dataframe to list of dataframes
    list_of_df.append(chunk)

# Concatenate all chunks into a single DataFrame
zillow_indicators_df = pd.concat(list_of_df, ignore_index=True)
zillow_indicators_df
```

```python
# Initialize an empty list to store chunks of Dataframes
list_of_df = []

# Read the CSV file in chunks
for chunk in pd.read_csv("../Datasets/ZILLOW_REGIONS.csv", chunksize=chunk_size):
    # add each dataframe to list of dataframes
    list_of_df.append(chunk)

# Concatenate all chunks into a single DataFrame
zillow_regions_df = pd.concat(list_of_df, ignore_index=True)
zillow_regions_df
```

```python
oh_regions = zillow_regions_df['region_id'].loc[(zillow_regions_df['region'].str.contains('OH;'))]

oh_data_df = zillow_data_df.loc[(zillow_data_df['region_id'].isin(oh_regions))]
oh_data_df.reset_index(inplace=True,drop=True)
oh_data_df

def split_data_to_csv(df):
    csv_limit = 1048000
    num_df_files = int((len(df))/csv_limit)

    #add one more file if the total files can hold all data 
    num_df_files = (num_df_files+1) if(len(df) > (num_df_files*csv_limit)) \
                                        else num_df_files
    
    print(f"files: {num_df_files} total: {num_df_files*csv_limit} and len: {len(df)}")

    for i in range(0, num_df_files):
        start_index = (i*csv_limit)
        end_index = (((i+1)*csv_limit)-1)

        print(f"{(i+1)}: {start_index} - {end_index}")
        df_file = df.iloc[start_index:end_index]
        df_file.to_csv("../Datasets/Data/Zillow_data_" + str(i+1) +".csv")

split_data_to_csv(oh_data_df)
```
