```python
import pandas as pd
import os
```

### Read and concat all the data from csv's to a single data frame

```python
files_to_read = os.listdir("../Datasets/Data/")

display(files_to_read)

read_files = []

for eachFile in files_to_read:
    if (eachFile.__contains__('data')):
        read_files.append(pd.read_csv("../Datasets/Data/" + eachFile))

oh_zillow_df = pd.concat(read_files, ignore_index=True)
oh_zillow_df
```

### Read and create dataframes for indicators and regions

```python
zillow_indicators_df = pd.read_csv("../Datasets/ZILLOW_INDICATORS.csv")
#display(zillow_indicators_df)

zillow_regions_df = pd.read_csv("../Datasets/ZILLOW_REGIONS.csv")
#display(zillow_regions_df)

#Extract only OH regions
oh_zillow_regions_df = zillow_regions_df.loc[zillow_regions_df['region'].str.contains('OH')].reset_index(drop=True)
display(oh_zillow_regions_df)
```
